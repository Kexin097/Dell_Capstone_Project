Import Modules

Data Visualisation - Explore data deeper

```python
# Explore Categorical Data

cat_columns = ['PaidBy_Region', 'Work_Country', 'Location', 'ESG_HES', 'Product_LOB', 'Queue Name', 'EPI_TITLE', 'Job_Title', 'Tenure_Group']

# Create 3x3 subplots
fig, axs = plt.subplots(3, 3, sharex=False, sharey=False, figsize=(20, 15))
fig.subplots_adjust(hspace=1.2)

# Creating counter to loop through cat_columns
counter = 0
for cat_column in cat_columns:
    value_counts = data[cat_column].value_counts()
    
    trace_x = counter // 3
    trace_y = counter % 3
    x_pos = np.arange(0, len(value_counts))
    
    axs[trace_x, trace_y].bar(x_pos, value_counts.values, tick_label = value_counts.index)
    
    axs[trace_x, trace_y].set_title(cat_column)
    
    for tick in axs[trace_x, trace_y].get_xticklabels():
        tick.set_rotation(90)
    
    counter += 1

plt.show()
```

Pre-Processing Data

```python
# Import LabelEncoder

le = preprocessing.LabelEncoder()

data['Product_LOB']=le.fit_transform(data['Product_LOB'])
data['Job_Title']=le.fit_transform(data['Job_Title'])
data['Tenure_Group']=le.fit_transform(data['Tenure_Group'])
```

```python
data.head(50)
```

Split Train and Test Set

```python
# Specifying columns 
X = data[['Product_LOB', 'Job_Title', 'Tenure_Group']]
y = data['Attr_Status'] 


# Splitting by stratifying by cluster name for equal representation of data  # test size=30%, train 70%

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

Model Building

# Gradient Boosting Classifier

```python
#Import Gradient Boosting Classifier model
from sklearn.ensemble import GradientBoostingClassifier

#Create Gradient Boosting Classifier
gb = GradientBoostingClassifier()

#Train the model using the training sets
gb.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = gb.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision
print("Precision:",metrics.precision_score(y_test, y_pred))

# Model Recall
print("Recall:",metrics.recall_score(y_test, y_pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

dt = DecisionTreeClassifier()
rm = RandomForestClassifier()
gnb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
for a,b in zip([dt,knn,rm,gnb],["Decision Tree", "Random Forest", "Naive Bayes", "KNN"]):
    a.fit(X_train,y_train)
    prediction=a.predict(X_train)
    y_pred=a.predict(X_test)
    score1=accuracy_score(y_train,prediction)
    score=accuracy_score(y_test,y_pred)
    msg1="[%s] training data accuracy is : %f" % (b,score1)
    msg2="[%s] test data accuracy is : %f" % (b,score)
    print(msg1)
    print(msg2)
```

```python
model_scores={'Decision tree':dt.score(X_test,y_test),
              'Random forest':rm.score(X_test,y_test),
              'Naive Bayes':gnb.score(X_test,y_test),
              'KNN classifier':knn.score(X_test,y_test)
             }
model_scores
```

Classification Report of Random Forest

```python
from sklearn.metrics import classification_report

rm_y_preds = rm.predict(X_test)

print(classification_report(y_test,rm_y_preds))
```

Classification Report of KNN

```python
from sklearn.metrics import classification_report

knn_y_preds = knn.predict(X_test)

print(classification_report(y_test, knn_y_preds))
```
